```python
# Install pandas into the current Python environment (run this cell once), then import it
import sys, subprocess
subprocess.check_call([sys.executable, "-m", "pip", "install", "--upgrade", "pip"])
subprocess.check_call([sys.executable, "-m", "pip", "install", "pandas"])
import pandas as pd
subprocess.check_call([sys.executable, "-m", "pip", "install", "matplotlib", "datasets"])
import matplotlib.pyplot as plt
from datasets import load_dataset
import sys, subprocess
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset["train"].to_pandas()
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
top_countries = df['job_country'].value_counts().head(6).index

df_job_country_salary = df.pivot_table(
    values = 'salary_year_avg',
    index = 'job_country',
    columns = 'job_title_short',
    aggfunc='median')

df_job_country_salary = df_job_country_salary.loc[top_countries]

job_title = ['Business Analyst', 'Data Analyst', 'Data Scientist']

df_job_country_salary = df_job_country_salary[job_title]

df_job_country_salary.plot(kind='bar', color=['pink', 'black', 'yellow'])
plt.title('Median Salary by Job Title and Country')
plt.xlabel('')
plt.ylabel('Median Salary (USD)')
plt.xticks(rotation=45, ha='right')
plt.show()
```
