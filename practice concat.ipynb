```python
data_scientist_df = df[df['job_title_short'] == 'Data Scientist']
data_analyst_df = df[df['job_title_short'] == 'Data Analyst']
combined_df = pd.concat([data_scientist_df, data_analyst_df], ignore_index=True)
combined_df
```

```python
df['job_posted_month']= df['job_posted_date'].dt.strftime('%b')
months = df['job_posted_month'].unique()
```

```python
months_dict = {month: df[df['job_posted_month'] == month] for month in months}
```

```python
df_q4 = pd.concat([months_dict['Oct'], months_dict['Nov'], months_dict['Dec']], ignore_index=True)
df_q4
```
