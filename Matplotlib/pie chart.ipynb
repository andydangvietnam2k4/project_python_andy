```python
# Install pandas into the current Python environment (run this cell once), then import it
import sys, subprocess
subprocess.check_call([sys.executable, "-m", "pip", "install", "--upgrade", "pip"])
subprocess.check_call([sys.executable, "-m", "pip", "install", "pandas"])
import pandas as pd
subprocess.check_call([sys.executable, "-m", "pip", "install", "matplotlib", "datasets"])
import matplotlib.pyplot as plt
from datasets import load_dataset
import sys, subprocess
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset["train"].to_pandas()
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
from ast import literal_eval
df['job_skills']= df['job_skills'].apply(lambda x: literal_eval(x) if pd.notna(x) else x)
```

```python
df['job_work_from_home'].value_counts().plot(kind='pie', startangle =90,autopct='%.1f%%')
plt.title('Job Work From Home Distribution')
```

```python
fig, ax = plt.subplots(1,3)

dict_column  = {
    'job_work_from_home': 'Work from Home',
    'job_no_degree_mention': 'Job Degree Req',
    'job_health_insurance': 'Health Insurance Offered',
}
for i, (column_name, column_title) in enumerate(dict_column.items()):
    ax[i].pie(df[column_name].value_counts(), startangle =90, autopct='%.1f%%',labels=['True','False'])
    ax[i].set_title(column_title)
plt.show()


#ax.pie(df['job_work_from_home'].value_counts())
```
